```python
import sympy as sym


m, c, k, t = sym.symbols('m c k t', real=True)

x = sym.Function('x')
```

```python
lhs = m * sym.Derivative(x(t), t, t) + c * sym.Derivative(x(t), t) + k * x(t)
lhs
```

```python
s = sym.dsolve(lhs, x(t),)

```

```python
s
```

```python
s1 = s.rhs
s1
```

```python
x0 = s1.subs({t:0})
x0
```

```python
s1.diff(t)
```

```python
v = s1.diff(t)
v
```

```python
v0 = sym.expand(v.subs({t:0}))
v0
```

```python
x_0, v_0 = sym.symbols('x_0 v_0', real=True)
```

```python
eq_init_0 = sym.Eq(x_0, x0)
eq_init_0
```

```python
eq_init_1 = sym.Eq(v_0, v0)
eq_init_1
```

```python
sym.solve([eq_init_0, eq_init_1])
```

```python
x0
```

```python
C1, C2 = x0.as_ordered_terms()
```

```python
sym.solve([eq_init_0, eq_init_1], (C1, C2))
```
